We think that data comes from

$Y_i = \alpha + \beta X_i + \varepsilon_i$


But data really comes from this

$Y_i = \alpha + \beta \log(X_i) + \varepsilon_i$



True DGP is not the same as the DGP we assumed when estimating

```python
R = 100
n = 1000
alpha = .3
beta = 1.5
const = np.ones([n,1])


param_list = []
for r in range(R):
    X = np.random.uniform(0,1,[n,1])
    err = np.random.normal(0,1,[n,1])
    Y = alpha + beta*np.log(X)+err
    # estimate the model
    exvar = np.concatenate([const,X],axis = 1)
    mod = sm.OLS(Y,exvar)
    res = mod.fit()
    param_list = param_list + [res.params[1]]
```

```python
np.mean(param_list)
```

True DGP is  the same as the DGP we assumed when estimating

```python
R = 100
n = 1000
alpha = .3
beta = 1.5
const = np.ones([n,1])


param_list = []
for r in range(R):
    X = np.random.uniform(0,1,[n,1])
    err = np.random.normal(0,1,[n,1])
    Y = alpha + beta*np.log(X)+err
    # estimate the model
    exvar = np.concatenate([const,np.log(X)],axis = 1)
    mod = sm.OLS(Y,exvar)
    res = mod.fit()
    param_list = param_list + [res.params[1]]
```

```python
np.mean(param_list)
```
